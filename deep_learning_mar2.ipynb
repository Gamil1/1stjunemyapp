```python
# library - deep learning
# tensorflow - Google - tensor means an array of data
## computational flow: x1.w1+x2.w2+....x1000w1000+b - loss | modify the set of weights
## modify the weights to reduce loss - optimizer
## with out keras: 2016: input, weights, bias, tf.matmul()- 100 lines of code
## keras - high level library built on top of tensorflow, less than 10 lines of code
## tf.keras - 2.x
# pytorch:
```

```python
import tensorflow as tf
import numpy as np # arrays of data - define my data
import matplotlib.pyplot as plt # plotting the charts
# age, experience and education
X = np.array([[22,2,1], # 30
             [25,3,1], #35
             [30,7,2], #50
             [35,10,2],
              [40,15,3]],
             dtype=np.float32) #float4, [accomodation of decimals - lesser the storage and ram]
# income
Y = np.array([30,35,50,60,80]).reshape(-1,1)
# (1,5)
model = tf.keras.Sequential([tf.keras.layers.Dense(1,input_shape=[3])]) #output
# Dense means fully connected layer
# (5,3).(3,1) = (5,1)
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.0001), loss='mse')
history = model.fit(X,Y,epochs=500,verbose=2)
```

```python
weights, bias = model.layers[0].get_weights()
print(f"'weights', {weights.flatten()}")
print("bias", bias)
```
